```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv(('C:\\Users\\Jatin\\Downloads\\company_sales_data.csv\\company_sales_data.csv'))

print(df)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(df['month_number'], df['total_profit'], linestyle='dotted', color='red', marker='o', markersize=8, markerfacecolor='red', label='Total Profit', linewidth=3)
plt.legend(loc='lower right')
plt.xlabel('Month Number')
plt.ylabel('Total Profit')
plt.title('Total Profit of All Months')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
for product in df.columns[2:7]:
    plt.plot(df['month_number'], df[product], label=product)

plt.legend()
plt.xlabel('Month Number')
plt.ylabel('Sold Units Number')
plt.title('Product Sales Data')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['month_number'], df['toothpaste'], color='blue')
plt.xlabel('Month Number')
plt.ylabel('Toothpaste Sales')
plt.title('Toothpaste Sales Data')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.bar(df['month_number'], df['bathingsoap'], color='green')
plt.xlabel('Month Number')
plt.ylabel('Bathing Soap Sales')
plt.title('Bathing Soap Sales Data')
plt.savefig('bathing_soap_sales.png')
```

```python
plt.figure(figsize=(10, 6))
plt.hist(df['total_profit'], bins=10, edgecolor='black')
plt.xlabel('Total Profit')
plt.ylabel('Frequency')
plt.title('Histogram of Total Profit')
plt.show()
```

```python
last_year_sales = df[df['month_number'] > 6].sum(numeric_only=True)[2:7]
plt.figure(figsize=(8, 8))
plt.pie(last_year_sales, labels=last_year_sales.index, autopct='%1.1f%%', startangle=90)
plt.title('Total Sale Data for Last Year')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


df = pd.read_csv("C:\\Users\\Jatin\\Downloads\\supermarket_sales - supermarket_sales.csv")

df
```

```python
max_quantity_duration = df.groupby('Date')['Quantity'].sum().idxmax()
plt.figure(figsize=(10, 6))
plt.plot(df.groupby('Date')['Quantity'].sum())
plt.title('Maximum Quantities Sold Over Time')
plt.xlabel('Date')
plt.ylabel('Quantity')
plt.show()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df['Month'] = df['Date'].dt.month
df['Hour'] = df['Time'].apply(lambda x: int(x.split(':')[0]))
plt.figure(figsize=(12, 6))
sns.barplot(x='Month', y='Quantity', hue='Gender', data=df.groupby(['Month', 'Hour', 'Gender'])['Quantity'].sum().reset_index())
plt.title('Distribution of Branch Sales Quantity per Hour (Monthly) with Gender Type')
plt.xlabel('Month')
plt.ylabel('Quantity')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.countplot(x='Payment', data=df)
plt.title('Payment Method Usage Count')
plt.xlabel('Payment Method')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.boxplot(x='Branch', y='Rating', data=df)
plt.title('Boxplot of Ratings by Branch')
plt.xlabel('Branch')
plt.ylabel('Rating')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.swarmplot(x='Customer type', y='Rating', data=df)
plt.title('Swarm Plot of Customer Type and Rating')
plt.xlabel('Customer Type')
plt.ylabel('Rating')
plt.show()
```

```python
sns.pairplot(df[['gross income', 'Quantity', 'Unit price', 'Rating']])
plt.suptitle('Pair Plot of Gross Income, Quantity, Unit Price, and Ratings')
plt.show()
```

```python
numeric_columns = df.select_dtypes(include=['float64', 'int64']).columns
df_numeric = df[numeric_columns]

correlation_matrix = df_numeric.corr()

plt.figure(figsize=(10, 8))
heatmap = plt.pcolor(correlation_matrix, cmap='coolwarm')
plt.colorbar(heatmap)

plt.xticks(range(len(correlation_matrix.columns)), correlation_matrix.columns, rotation='vertical')
plt.yticks(range(len(correlation_matrix.columns)), correlation_matrix.columns)

plt.title('Correlation Heatmap (matplotlib)')
plt.show()
```
